```python
df["order_id"] = df["order_id"].astype("Int64")
df["customer_id"] = df["customer_id"].astype("Int64")
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
```

```python
df.dtypes
```

```python
df = df[(df["price"] >= 0) & (df["quantity"] >= 0) & (df["total_amount"] >= 0)]
```

```python
text_cols = ["product","category","city","region","store",
             "sales_channel","size","payment_method"]

for c in text_cols:
    df[c] = df[c].str.strip()
```

```python
df
```

```python
num_cols = df.select_dtypes(include="number").columns


num_cols = [c for c in num_cols if c not in ["order_id","customer_id"]]

df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
df
```

```python
cat_cols = df.select_dtypes(include="object").columns
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
df = df[df["price"] >= 0]
df = df[df["quantity"] >= 0]
```

```python
df[df["rating"] > 5]
df["rating"] = df["rating"].clip(0, 5)
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
```

```python
df.to_csv("puma_cleaned.csv", index=False)
```
